# Pandas는 array와 다른 [][] 형태
ex) data['0']: row가 아닌 index(세로) 아래로 내려가는 형태
    data['Top1']: row가 아닌 아래로 내려가는 형태

```python
train = df[df['Date'] < '20150101']
test = df[df['Date'] > '20141231']
```

```python
train = df[(df['Date'] > '20100101') & (df['Date'] < '20150101')]
test = df[df['Date'] > '20141231']
```

```python
train.head()
```

```python
data = train.iloc[:, 2:27]
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

new_Index = [str(i) for i in range(25)]
data.columns = new_Index
data.head(5)
```

```python
for index in new_Index:
    data[index]=data[index].str.lower()
data.head()
```

```python
' '.join(str(x) for x in data.iloc[0, 0:25])
```

```python
headlines = []
for row in range(0, len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row, 0:25]))
```

```python
headlines[1]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

# fit과 fit_transform의 차이

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(traindataset, train['Label'])
```

```python
test_headlines = []
for row in range(0, len(test.index)):
    test_headlines.append(' '.join(str(x) for x in test.iloc[row, 0:25]))
testdataset = countvector.transform(test_headlines)
predictions = randomclassifier.predict(testdataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['Label'],predictions)
print(matrix)
score=accuracy_score(test['Label'],predictions)
print(score)
report=classification_report(test['Label'],predictions)
print(report)
```

# got about 85% accuracy rate
