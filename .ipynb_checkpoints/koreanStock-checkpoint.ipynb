# Training news for 4 years & Test news for 1 year

```python
train = df[df['date'] < '2019-01-01']
test = df[df['date'] > '2018-12-31']
```

```python
train = df[df['date'] < '2018-05-01']
test = df[df['date'] > '2018-04-30']
```

```python
train.shape
```

```python
test.shape
```

```python
data = train.iloc[:, 2:42]
```

```python
#data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)
data.replace("[^가-힣a-zA-Z]"," ",regex=True, inplace=True)
new_Index = [str(i) for i in range(40)]

data.columns = new_Index
data.head(5)
```

```python
' '.join(str(x) for x in data.iloc[0, 0:40])
```

```python
headlines = []
for row in range(0, len(data.index)):
    headlines.append(' '.join(str(x) for x in data.iloc[row, 0:40]))
```

```python
headlines[0]
```

# Training model

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
countvector = CountVectorizer(ngram_range=(2,2))
traindataset = countvector.fit_transform(headlines)
```

```python
randomclassifier = RandomForestClassifier(n_estimators=200, criterion='entropy')
randomclassifier.fit(traindataset, train['label'])
```

```python
test_headlines = []
for row in range(0, len(test.index)):
    test_headlines.append(' '.join(str(x) for x in test.iloc[row, 0:40]))
testdataset = countvector.transform(test_headlines)
predictions = randomclassifier.predict(testdataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(test['label'],predictions)
print(matrix)
score=accuracy_score(test['label'],predictions)
print(score)
report=classification_report(test['label'],predictions)
print(report)
```
